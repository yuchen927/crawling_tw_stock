```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import numpy as np
```

```python
# Use HiStock to crawl Taiwan stock data
url = "https://histock.tw/stock/rank.aspx?p=all"
user_agent = UserAgent()
headers = {'user-agent': user_agent.random}
```

```python
# crawl webpage html information
res = requests.get(url, headers = headers)
soup = BeautifulSoup(res.text, 'lxml')
soup
```

```python
# Find out where the table is stored 
# all prices info is at id = “CPHB1_gv”
tmp = soup.select_one('#CPHB1_gv')
```

```python
df = pd.read_html(tmp.prettify())
```

```python
type(df)
```

```python
# reshape the list 
df = np.reshape(df, (2152,13))
```

```python
df = pd.DataFrame(np.array(df), columns= ['code', 'name', 'price', 'change', 'change_%', 'week_change_%', 'amplitude', 'open', 'highest', 'lowest', 'close_ystdy', 'volume', 'volume_amount_yi'])
```

```python
df.head()
```

```python
df.to_csv('tw_stock_0207.csv', encoding='utf_8_sig')
```
